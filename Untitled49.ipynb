```python
# Librerías para manipulación de datos
import numpy as np
import pandas as pd

# Librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Librerías para Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Modelos de Machine Learning
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Evaluación de modelos
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Cargar los datos desde el archivo CSV
cardio_data = pd.read_csv('DataSetCardio.csv')
```

```python
# Mostrar las primeras filas del dataset para entender su estructura
print("Primeras 5 filas del dataset:")
cardio_data.head()
```

```python
# Mostrar información sobre el dataset
print("\nInformación del dataset:")
cardio_data.info()
```

```python
# Mostrar estadísticas descriptivas
print("\nEstadísticas descriptivas:")
cardio_data.describe()
```

```python
# Verificar si hay valores nulos
print("\nValores nulos por columna:")
cardio_data.isnull().sum()
```

```python
# Eliminar filas donde 'trestbps' es nulo
cardio_data = cardio_data[cardio_data["trestbps"].notnull()]

# Rellenar los nulos de 'chol' con la media
cardio_data["chol"] = cardio_data["chol"].fillna(cardio_data["chol"].mean())

# Rellenar los nulos de 'fbs' con la moda
cardio_data["fbs"] = cardio_data["fbs"].fillna(cardio_data["fbs"].mode()[0])

# Verificar que ya no hay nulos
print("Valores nulos:\n")
print(cardio_data.isnull().sum())
```

```python
# ¿Cuántos pacientes hay en el dataset?
total_patients = len(cardio_data)
print(f"Total de pacientes en el dataset: {total_patients}")
```

```python
# ¿Cuántos pacientes presentan enfermedad cardíaca?
conteo = cardio_data['target'].value_counts().rename(index={1: 'Presencia de enfermedad', 0: 'Ausencia de enfermedad'})
print(conteo)
```

```python
# ¿Cuáles son las columnas disponibles en el dataset?
columnas = cardio_data.columns
print("\nColumnas disponibles:")
print(list(columnas))
```

```python
# ¿Qué tipo de datos contiene cada columna?
tipos_datos = cardio_data.dtypes
print("\nTipos de datos por columna:")
print(tipos_datos)
```

```python
# Gráfica de caja y bigotes
sns.boxplot(x='target', y='chol', data=cardio_data)
plt.title('Boxplot de colesterol según presencia de enfermedad')
plt.xlabel('Presencia de enfermedad (0 = No, 1 = Sí)')
plt.ylabel('Colesterol')
plt.show()
```

```python
# Gráfica de violín
sns.violinplot(x='target', y='chol', data=cardio_data)
plt.title('Violin plot de colesterol según presencia de enfermedad')
plt.xlabel('Presencia de enfermedad (0 = No, 1 = Sí)')
plt.ylabel('Colesterol')
plt.show()
```

```python
# Análisis de correlación
plt.figure(figsize=(12, 10))
# Seleccionamos solo columnas numéricas para la matriz de correlación
numeric_cols = cardio_data.select_dtypes(include=['int64', 'float64']).columns
correlation = cardio_data[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.tight_layout()
plt.show()
```
